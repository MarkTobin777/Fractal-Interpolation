```python
from random import randint
import matplotlib.pyplot as plt
```

```python
x,y = [0],[0]
x1,x2,x3 = [],[],[]
y1,y2,y3 = [],[],[]

# change range to generate different numbers of points

for i in range(10000):
    map = randint(1,3)

    xi = x[0]
    yi = y[0]
    
    if map == 1:
        xi = (x[i])/2
        yi = (y[i])/2
        
        x.append(xi);y.append(yi)
        x1.append(xi);y1.append(yi)

    elif map == 2:
        xi = (x[i]+1.5)/2
        yi = (y[i]+3)/2
        
        x.append(xi);y.append(yi)
        x2.append(xi);y2.append(yi)

    else:
        xi = (x[i]+3)/2
        yi = (y[i])/2
        
        x.append(xi);y.append(yi)
        x3.append(xi);y3.append(yi)

plt.scatter(x1,y1, s = 1, color = "Black") # change colours here
plt.scatter(x2,y2, s = 1, color = "Black")
plt.scatter(x3,y3, s = 1, color = "Black")
plt.axis("off")
plt.savefig('sierpinski3.jpg')
plt.show()
```

```python
# barnsley fern

x1,x2,x3,x4 = [],[],[],[]
y1,y2,y3,y4 = [],[],[],[]
x,y = [0],[0]

xi = x[0]
yi = y[0]

c=0

for i in range(100000):
    map = randint(1,100)

    if map == 1:
        xi = 0
        yi = 0.16*y[c]

        x.append(xi);y.append(yi)
        x1.append(xi);y1.append(yi)

    elif map >= 2 and map <= 84:
        xi = 0.85*x[c] + 0.02*y[c]
        yi = -0.02*x[c] + 0.85*y[c] + 1.6

        x.append(xi);y.append(yi)
        x2.append(xi);y2.append(yi)

    elif map >= 85 and map <= 90:
        xi = 0.2*x[c] - 0.26*y[c]
        yi = 0.23*x[c] + 0.22*y[c] + 1.6

        x.append(xi);y.append(yi)
        x3.append(xi);y3.append(yi)

    else:
        xi = -0.15*x[c] + 0.28*y[c]
        yi = 0.26*x[c] + 0.24*y[c] + 0.44

        x.append(xi);y.append(yi)
        x4.append(xi);y4.append(yi)

    c = c+1

plt.scatter(x1,y1, s = 0.01, color = "Darkgreen")
plt.scatter(x2,y2, s = 0.01, color = "Darkgreen")
plt.scatter(x3,y3, s = 0.01, color = "Darkgreen")
plt.scatter(x4,y4, s = 0.01, color = "Darkgreen")
plt.axis("off")
plt.savefig('fern2.jpg')
plt.show()
```

```python
## sierpinski carpet with 

x1,x2,x3,x4,x5,x6,x7,x8 = [],[],[],[],[],[],[],[]
y1,y2,y3,y4,y5,y6,y7,y8 = [],[],[],[],[],[],[],[]


x,y = [0],[0]
xi = x[0];yi = y[0]

for i in range(50000):
    map = randint(1,800)

    xi = x[0]
    yi = y[0]
    

    if map < 100:
        xi = (x[i])/3
        yi = (y[i])/3
        
        x.append(xi);y.append(yi)
        x1.append(xi);y1.append(yi)


    elif map >= 100 and map < 300:
        xi = (x[i])/3 +(1/3)
        yi = (y[i])/3
        
        x.append(xi);y.append(yi)
        x2.append(xi);y2.append(yi)

    elif map >= 200 and map < 250:
        xi = (x[i])/3  +(2/3)
        yi = (y[i])/3
        
        x.append(xi);y.append(yi)
        x3.append(xi);y3.append(yi)
    
    elif map >=250 and map < 300:
        xi = (x[i])/3 
        yi = (y[i])/3  +(2/3)
        
        x.append(xi);y.append(yi)
        x4.append(xi);y4.append(yi)

    elif map >= 300 and map < 500:
        xi = (x[i])/3  +(1/3)
        yi = (y[i])/3  +(2/3)
        
        x.append(xi);y.append(yi)
        x5.append(xi);y5.append(yi)

    elif map > 500 and map <= 575:
        xi = (x[i])/3 +(2/3)
        yi = (y[i])/3  +(2/3)
        
        x.append(xi);y.append(yi)
        x6.append(xi);y6.append(yi)

    elif map > 575 and map <= 700:
        xi = (x[i])/3
        yi = (y[i])/3  +(1/3)
        
        x.append(xi);y.append(yi)
        x7.append(xi);y7.append(yi)

    else:
        xi = (x[i])/3  +(2/3)
        yi = (y[i])/3  +(1/3)
        
        x.append(xi);y.append(yi)
        x8.append(xi);y8.append(yi)

ax = plt.plot()
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')

plt.scatter(x1,y1, s = 0.5, color = "Black")
plt.scatter(x2,y2, s = 0.5, color = "Black")
plt.scatter(x3,y3, s = 0.5, color = "Black")
plt.scatter(x4,y4, s = 0.5, color = "Black")
plt.scatter(x5,y5, s = 0.5, color = "Black")
plt.scatter(x6,y6, s = 0.5, color = "Black")
plt.scatter(x7,y7, s = 0.5, color = "Black")
plt.scatter(x8,y8, s = 0.5, color = "Black")

plt.axis("off")
plt.savefig("carpet1.pdf")
plt.show()
```
